```python
def split(data,p):
    a=[]
    b=[]
    n=data.shape[0]
    
    for i in range(n):
        x=random.random()
        if x<=p:
            a.append(i)
        else:
            b.append(i)
    return data.iloc[a],data.iloc[b]        
```

```python
test,train=split(df,0.3)
```

```python
#variables used in model
vars=['Age','BMI','DiabetesPedigreeFunction','Glucose']
```

```python
# The code in cell 18 uses 'xrange', which is not available in Python 3.
# Replace 'xrange' with 'range' and update print statements for Python 3 compatibility.

def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))

def hypothesis(x, coef):
    h = coef[0]
    for i in range(1, len(coef)):
        h += coef[i] * x[i-1]
    return sigmoid(h)

def gradient_decent(x, y, coef):
    n1 = x.shape[0]
    b = np.array([0.0 for i in range(len(coef))], dtype=np.float64)
    b1 = np.array([[0.0 for i in range(len(coef))] for j in range(len(coef))])

    for i in range(n1):
        h = hypothesis(x.iloc[i], coef)
        b[0] += (h - float(y.iloc[i])) * (1.0 / float(n1))
        c1 = [[1.0]]

        for j in range(1, len(coef)):
            b[j] += (1.0 / float(n1)) * (h - float(y.iloc[i])) * x.iloc[i, j-1]
            c1.append([x.iloc[i, j-1]])

        c1 = np.array(c1, dtype=np.float64)
        c = c1.reshape((1, len(coef)))
        c1 = c1.dot(c)
        c1 = c1 * h * (1 - h) * (1.0 / float(n1))
        b1 += c1

    H = np.linalg.inv(b1)
    H = H.dot(b)
    coef -= H
    return coef

def LogisticRegression(x, y, iterations):
    coef = np.array([0 for i in range(x.shape[1] + 1)], dtype=np.float64)
    for iteration in range(iterations):
        coef = gradient_decent(x, y, coef)
    return coef
```

```python
#training data and calculating coefficients 
coef=LogisticRegression(train[vars],train['Outcome'],10)
```

```python
#function to predict dataset
def predict(x,y,coef):
    accuray=0
    for i in range(x.shape[0]):
        pred=hypothesis(x.iloc[i],coef)
        if pred>=0.4:
            o=1
        else:
            o=0
        if o==y.iloc[i]:
            accuray+=1

    print ('Accuracy :',accuray/float(x.shape[0])*100.0)        
```

```python
#Accuracy 
predict(test[vars],test['Outcome'],coef)
```

```python
from IPython.display import display, HTML

# Enhanced accuracy display with color and formatting

def predict_with_style(x, y, coef):
    correct = 0
    for i in range(x.shape[0]):
        pred = hypothesis(x.iloc[i], coef)
        o = 1 if pred >= 0.4 else 0
        if o == y.iloc[i]:
            correct += 1
    accuracy = correct / float(x.shape[0]) * 100.0
    html = f"""
    <div style="background: linear-gradient(90deg, #43cea2, #185a9d); padding: 20px; border-radius: 10px; text-align: center;">
        <h2 style="color: white; font-weight: bold; margin: 0;">Model Accuracy</h2>
        <p style="font-size: 2em; color: #ffd700; margin: 10px 0 0 0;">{accuracy:.2f}%</p>
    </div>
    """
    display(HTML(html))

# Usage
predict_with_style(test[vars], test['Outcome'], coef)
```
